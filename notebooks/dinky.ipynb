# Impactful delays

Two main questions:
* Is the Dinky ever delayed such that one misses an outbound train at Princeton Junction?
* How often does a inbound train get delayed such that one misses the Dinky?

```python
# if I go down the df sorted by time, and I get consecutive to/from's of Princeton Junction...
```

```python
princeton = princeton.sort_values("expected")
```

```python
moving_princeton = princeton[princeton["stop_sequence"] != 1.0]
```

```python
# try shift?
```

```python
moving_princeton["shifted_from"] = moving_princeton["from"].shift(1)
```

```python
moving_princeton["shifted_to"] = moving_princeton["to"].shift(1)
```

```python
moving_princeton["shifted_expected"] = moving_princeton["expected"].shift(1)
```

```python
moving_princeton["shifted_time"] = moving_princeton["time"].shift(1)
```
